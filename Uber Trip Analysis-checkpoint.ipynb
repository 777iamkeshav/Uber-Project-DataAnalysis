```python
df['date'] = pd.to_datetime(df['date'])
if 'trips' not in df.columns:
    df['trips'] = 1
```

```python
daily = df.groupby('date')['trips'].sum().reset_index().sort_values('date')
daily = daily.set_index('date')
```

```python
# Quick look
print("Daily series length:", len(daily))
display(daily.head())
```

```python
ts = daily.copy()
ts['y'] = ts['trips']
```

```python

# lags
ts['lag_1'] = ts['y'].shift(1)
ts['lag_7'] = ts['y'].shift(7)
ts['lag_14'] = ts['y'].shift(14)
```

```python
# rolling stats
ts['roll_7_mean'] = ts['y'].shift(1).rolling(window=7).mean()
ts['roll_14_mean'] = ts['y'].shift(1).rolling(window=14).mean()
ts['roll_7_std'] = ts['y'].shift(1).rolling(window=7).std().fillna(0)
```

```python
# date features
ts['dayofweek'] = ts.index.dayofweek
ts['is_weekend'] = ts['dayofweek'].isin([5,6]).astype(int)
ts['month'] = ts.index.month
ts['day'] = ts.index.day
```

```python
# drop rows with NaNs due to lags
ts = ts.dropna().copy()
print("After feature creation, rows:", len(ts))
```

```python
n = len(ts)
train_end = int(n * 0.8)
train = ts.iloc[:train_end]
test = ts.iloc[train_end:]
```

```python

X_cols = ['lag_1','lag_7','lag_14','roll_7_mean','roll_14_mean','roll_7_std','dayofweek','is_weekend','month','day']
X_train = train[X_cols]
y_train = train['y']
X_test = test[X_cols]
y_test = test['y']
```

```python
print("Train rows:", X_train.shape[0], "Test rows:", X_test.shape[0])
```

```python
df['lag_1'] = df['trips'].shift(1)
df = df.dropna() 
```

```python
X = df[['lag_1']]  # include lag_1 in features
y = df['trips']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np

y_pred_baseline = X_test['lag_1'].values
mae_base = mean_absolute_error(y_test, y_pred_baseline)
rmse_base = np.sqrt(mean_squared_error(y_test, y_pred_baseline))

print(f"Baseline (lag-1) MAE: {mae_base:.2f}, RMSE: {rmse_base:.2f}")
```

```python
lgb_train = lgb.Dataset(X_train, label=y_train)
params = {
    'objective': 'regression',
    'metric': 'rmse',
    'verbosity': -1,
    'random_state': 42
}
```

```python
# train
model = lgb.train(params, lgb_train, num_boost_round=100)
```

```python
# predict
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

# Predictions
y_pred = model.predict(X_test)

# MAE
mae = mean_absolute_error(y_test, y_pred)

# RMSE (manually taking the square root if 'squared' not supported)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

print(f"LightGBM MAE: {mae:.2f}, RMSE: {rmse:.2f}")

```
